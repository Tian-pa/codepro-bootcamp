# Diccionarios

##### Palabra: significado  ---> humanamente
##### clave: valor ---> programacion
##### "nombre": "Fulana"
##### Cada valor en el diccionario es sepadara por comas y todo se encuentra dentro de dos llaves del tipo {}
Se puede acceder a los datos/valor de un diccionario con nombre_de_diccionario['nombre']

Para cambiar o editar el valor de una clave de uno de los elementos del diccionario es:
nombre_de_diccionario = 'nombre'
nombre_de_diccionario['nombre'] = 'manchuca garcia'

Para que nos devuelva todas las claves del diccionario es nombre_de_diccionario.keys()
Para que nos devuelva todas las valores del diccionario es nombre_de_diccionario.values()
Para que nos devuelva todas las claves y valores del diccionario es nombre_de_diccionario.items()

Para anhadir mas datos al diccionario:
nombre_de_diccionario['nueva_clave'] = 'valor_nueva_clave'

Para eliminar el ultimo elemento del diccionario es:
nombre_de_diccionario.popitem()

Para eliminar datos por clave:
del nombre_de_diccionario['apellido']

Los diccionarios son iterables:
for valor in nombre_de_diccionario.values():
    print(valor)

for claves in nombre_de_diccionario.keys():
    print(claves)

for valor, claves in nombre_de_diccionario.items():
    print(valor, claves)

# Definiendo m√©todos

```python
class Superheroe:
    # Atributo de clase
    especie = "superhumanos"

    # Definir m√©todo constructor para definir sus atributos de instancia
    def __init__(self, nombre):
        self.nombre = nombre

    # M√©todo
    def rayos_laser(self):
        print("tiro rayos laser rojos")

    def volar(self, km):
        print(f'Estoy volando a {km} km por hora')

    def alzar_objetos(self, objeto, kg):
        print(f'Puedo alzar un {objeto} de {kg} kg')

# Instanciar objetos
superheroe1 = Superheroe('Batman')
print(superheroe1.volar(120))

superheroe2 = Superheroe('Flash')
superheroe2.alzar_objetos("trenes", "5")
```

```python
class Coche:
    tipo = "vehiculo"

    # Atributos de instancia
    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion
    
    # Definir otras acciones de la clase
    def acelerar(self, metros):
        self.posicion += metros                                   # metros es propio del m√©todo, por ende, no se utiliza el self
        print(f'El vehiculo modelo {self.modelo} acelera y avanza {metros}. Nueva posici√≥n: {self.posicion}')

    def girar(self, direccion):
        print(f"El {self.modelo} avanza y gira a la direccion {direccion}")

coche1 = Coche("Toyota")
coche1.acelerar(1)
coche1.girar("izquierda")
```

# Herencia omaigoto :o

```python
class AnimalDeGranja:
    # Atributos de clase:
    tipo = "animal de granja"

    # Atributos de instancia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    # M√©todos (Acciones de la clase)
    def hablar(self):
        print(f"Hablo y digo {self.sonido}")

    def presentarse(self):
        print(f"Hola, soy {self.nombre}")

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevos(self):
        print(f"Estoy poniendo {self.huevos_por_dia} huevos por d√≠a")

class Vaca(Gallina):
    patas = 4

    def mujir(self):
        print("Muuu")

gallina1 = Gallina("Hortencia", "Marron", "Cocorocooo")
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca("Lola", "Blanco y negro", "Muuu")
vaca.hablar()
vaca.mujir()
```

Observaci√≥n: Editar las clases hijas para que no se inicialicen.

Challenge 4 ü§∫ - 20 min
Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python
# Clase Heroe
class Heroe:
    # Atributos de la clase heroe:
    atributo_extra = "vitalidad"

    # Atributos de la instancia
    def __init__(self, nombre, arma):
        self.nombre = nombre
        self.arma = arma

    # M√©todo de la clase Heroe
    def desenvaine(self):
        print(f"Desenvaina su {self.arma}")

# Clase Vikingo
class Vikingo:
    # Atributo de la clase Vikingo:
    atributo_extra = "furia"

    # Atributos de la instancia
    def __init__(self, nombre, arma = "hacha"):
        self.nombre = nombre
        self.arma = arma

    # M√©todo de la clase Vikingo
    def ataque_hacha(self):
        print(f"Ataca con su {self.arma}")

# Instanciando el objeto
heroe1 = Heroe("Odiseo", "espada")
# Impresi√≥n de la frase del h√©roe
print(f"Mi nombre es {heroe1.nombre} y mi tengo {heroe1.atributo_extra}")
#Llamada al m√©todo desenvaine
heroe1.desenvaine()

# Instanciando el objeto
berserker1 = Vikingo("Thorfinn")
# Impresi√≥n de la frase del Vikingo
print(f"\nMi nombre es {berserker1.nombre} y mi tengo {berserker1.atributo_extra}")
#Llamada al m√©todo ataque_hacha
berserker1.ataque_hacha()
```
