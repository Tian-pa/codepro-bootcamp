### Challenge ü§∫ - 15m
### Escribir un programa que almacene una contrase√±a en una variable, y pregunte al usuario por la contrase√±a hasta que introduzca la contrase√±a correcta.

### Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
# Challenge 11.

# Variables.
contrase√±a = "PeNgUiN"
# Inicia el bucle.
while True:
    # Pedimos al usuario que ingrese la contrase√±a.
    ingreso_contrase√±a = input("Ingrese la contrase√±a: ")
    # En el caso de que coincida la contrase√±a ingresada con la contrase√±a real.
    if ingreso_contrase√±a == contrase√±a:
        # El ciclo termina
        break
    # En el caso de que la contrase√±a ingresada no coincida con la contrase√±a real.
    else:
        # Se le pide al usuario que ingrese nuevamente la contrase√±a.
        print("Contrase√±a incorrecta, int√©ntelo nuevamente.")
# Mensaje secreto ;)
print("Contrase√±a correcta\n¬°A seguir codeando ping√ºinito!")
```

Un iterador es una variable temporal

### Challenge ü§∫ - 10m
##### Escribir un programa que itere en un rango de 10 veces. Por cada iteracion multiplicar el iterador por dos e imprimir el resultado. 

```python
# Challenge 13.

# For con un rango de 10.
for iterador in range(10):
    # Mostramos el numero de ciclo y el valor del iterador.
    print(f'{iterador + 1 }¬∫ Vuelta, el iterador tiene el valor: {iterador}')
    # Mostramos el valor del iterador, multiplicado por dos.
    print("Multiplicado por dos da: ", iterador * 2, "\n")
```

```python
# Challenge 14 -- PRIMER CHALLENGE QUE NO PUDE HACER A TIEMPO :C (edit: agregaron m√°s tiempo, as√≠ que s√≠ pude resolverlo a tiempo!!).

# Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene
# 6 letras o m√°s.

# Creamos la lista de marcas
lista_de_marcas = ["dove", "nike", "puma", "nissan", "asus"]

# Creamos el for que tendr√° como rango la lista de marcas
for iterador in lista_de_marcas:
    # Imprime el elemento de la lista que tenga como √≠ndice el valor del iterador
    print("La palabra de la lista es: ", iterador)
    # 
    print("La cantidad de letras es: ", len(iterador))
    if len(iterador) >= 6:
        print("¬°¬°Esta palabra de la lista tiene seis letras o m√°s!!\n")
    else:
        print("Esta palabra de la lista no tiene las seis letras m√≠nimas\n")
```

### Challenge ü§∫ - 10m
### Solicitar al usuario que ingrese una frase.

### Imprimir un mensaje cada vez que se encuentre una letra "o".

```python
# Challenge 15

frase = input("Por favor, ingrese una frase: ")
print("La frase ingresada fue: ", frase, "\n")

for letra in frase:
    if letra == "o":
        print(f'¬°Se ha encontrado una letra o!')
```

```python
# Una soluci√≥n que encontr√© en internet jeje, para poder 

# 1. Inicializar el contador ANTES del bucle
contador = 0

# 2. Definir el bucle for (por ejemplo, para iterar 5 veces)
for i in range(5):
  # Acci√≥n dentro del bucle (opcional, aqu√≠ solo imprimimos)
  print(f"Esta es la iteraci√≥n n√∫mero {i+1}")

  # 3. Incrementar el contador en cada paso
  contador += 1 # o contador = contador + 1

# 4. Mostrar el resultado final despu√©s del bucle
print(f"El bucle se ejecut√≥ {contador} veces.")
```
